# Bar plot of MAE of the different model

```python
import numpy as np
import matplotlib.pyplot as plt


# creating the dataset
data = {'Benchmark':351.26564067758295, 'KNN':228.93482320977935, 'Linear Regression':411.2606036142575}
courses = list(data.keys())
values = list(data.values())

fig = plt.figure(figsize = (10, 5))

# creating the bar plot
plt.bar(courses, values, color ='maroon',
		width = 0.4)

plt.xlabel("Type of model")
plt.ylabel("MAE")
plt.title("MAE of different model (train)")
plt.show()
```

Linear Regression Model has the highest MAE with least MAE of KNN (train).
